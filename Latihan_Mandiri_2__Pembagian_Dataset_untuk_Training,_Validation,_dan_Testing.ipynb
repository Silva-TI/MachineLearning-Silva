```python
import pandas as pd
import numpy as np
```

```python
# memasukan dataset
df = pd.read_csv("day.csv")
total_size = len(df)
```

```python
# pembagian rasio sesuai soal
rasio_testing = 0.20
rasio_validation = 0.10
rasio_training = 0.80
```

```python
# Hitung jumlah baris
jumlah_testing = int(np.ceil(total_size * rasio_testing))
jumlah_training = int(np.ceil(total_size * rasio_training))
jumlah_validation = int(np.ceil(jumlah_training * rasio_validation))
```

```python
# A. DATA TESTING (20%)
df_testing = df.head(jumlah_testing)
df_testing
```

```python
# B. DATA VALIDATION (10%)
df_validation = df.head(jumlah_validation)
df_validation
```

```python
# C. DATA TRAINING (80%)
df_training = df.head(jumlah_training)
df_training
```
